### [A4] ¿Cuánto tiempo esperan las personas hasta hacer otro pedido? Comenta sobre los valores mínimos y máximos.

```python
orders_prior= orders['days_since_prior_order'].value_counts().sort_index()
orders_prior.index= orders_prior.index.astype(int)
orders_prior.plot(
    title='Días transcurridos para su proxima compra', 
    xlabel='tiempo de espera(días)', 
    ylabel='Número de pedidos',
    kind='bar',
    color='c',
)
plt.show()
min_value=orders_prior.min()
min_days=orders_prior[orders_prior== min_value].index.tolist()
max_value=orders_prior.max()
max_days=orders_prior[orders_prior== max_value].index.tolist()

print(f'el número minimo de pedidos es {min_value},y ocurre a los:{min_days} días')
print(f'el número máximo de pedidos es {max_value},y ocurre a los:{max_days} días')
```

Un gran número de pedidos (51.337) se realizan a los 30 días de su anterior compra esto se puede deber a factores como los días de pago de nomina, entre otros, por otro lado transcurridos 26 días desde su anteror compra se encentra el escenario con menor número de pedidos (2.640).En conclusión, Las pedidos se hacen en su mayor medida los primeros 8 días despues de su primera compra o hasta cumplir los 30 días.

### [C1] ¿Cuántos artículos compran normalmente las personas en un pedido? ¿Cómo es la distribución?

```python
normal_quantity_orders=order_products.groupby('order_id')['product_id'].count()
normal_quantity_orders.value_counts().sort_index().plot(title='¿Cuántos artículos compran normalmente las personas en un pedido?',
                           kind='bar',
                           xlabel='Cantidad de articulos',
                           ylabel='Frecuencia', figsize=[12,10])
plt.show()
min_value=normal_quantity_orders.min()
max_value=normal_quantity_orders.max()
avg_value=normal_quantity_orders.mean()
print(f"Las personas compran minímo {min_value} articulos")
print(f"Las personas compran máximo {max_value} articulos")
print(f"Las personas compran en promedio {avg_value:.0f} articulos")
```

### [C2] ¿Cuáles son los 20 principales artículos que vuelven a pedirse con mayor frecuencia (muestra sus nombres e IDs de los productos)?

```python
products_with_reordered = order_products[order_products['reordered'] == 1]

top_products_with_reordered = products_with_reordered['product_id'].value_counts().head(20)
top_products = products[products['product_id'].isin(top_products_with_reordered.index)]
top_products = top_products[['product_id', 'product_name']]
top_products = top_products.merge(top_products_with_reordered.to_frame(), left_on='product_id', right_index=True)
top_products= top_products.rename(columns={'product_id_y':'Num_reorders'}).drop(columns=['product_id_x'])
top_products=top_products.sort_values('Num_reorders', ascending=False)
print(top_products)
top_products.set_index("product_name")["Num_reorders"].plot(kind="bar", figsize=[12, 6], 
                        color="m", title="Top 20 de productos que mas se vuelven a pedir",
                        ylabel='Número de reordenamientos', xlabel= 'Productos')
plt.gca().invert_xaxis()
plt.show()
```

Los productos mayor vendidos, vuelven a ser pedidos en su siguiente compra es decir que los clientes esan satisfechos con ellos. y podrian ser deterinantes en la decisión de compra para la mayoria de clientes.
